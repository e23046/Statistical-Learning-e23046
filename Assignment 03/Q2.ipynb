## Derivation of the Conditional Distribution $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

We want to show that the conditional distribution $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$ is also a multivariate normal distribution with specific mean $\boldsymbol{\mu}^*$ and covariance $\boldsymbol{\Sigma}_{X\mid Y}$.

The conditional density is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$.
Since $f_{\mathbf{Y}}(\mathbf{y})$ is the marginal density of $\mathbf{Y}$, it is a constant with respect to $\mathbf{x}$. Therefore, the conditional density is proportional to the joint density:

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) \propto f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})$$

We know the joint density is:
$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right)$$

Let's focus on the exponent term, as the normalization constant will be determined later. We can express $\mathbf{z} = \begin{bmatrix} \mathbf{x} \\ \mathbf{y} \end{bmatrix}$ and $\boldsymbol{\mu} = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix}$. Also, let $\boldsymbol{\Sigma}^{-1}$ be partitioned as:

$$\boldsymbol{\Sigma}^{-1} = \begin{bmatrix} \boldsymbol{\Lambda}_{XX} & \boldsymbol{\Lambda}_{XY} \\ \boldsymbol{\Lambda}_{YX} & \boldsymbol{\Lambda}_{YY} \end{bmatrix}$$

From the properties of partitioned matrices, we have:

*   $\boldsymbol{\Lambda}_{XX} = (\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX})^{-1} = \boldsymbol{\Sigma}_{X\mid Y}^{-1}$
*   $\boldsymbol{\Lambda}_{YY} = (\boldsymbol{\Sigma}_{YY} - \boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{XX}^{-1}\boldsymbol{\Sigma}_{XY})^{-1}$
*   $\boldsymbol{\Lambda}_{XY} = -\boldsymbol{\Sigma}_{XX}^{-1}\boldsymbol{\Sigma}_{XY}(\boldsymbol{\Sigma}_{YY} - \boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{XX}^{-1}\boldsymbol{\Sigma}_{XY})^{-1} = -\boldsymbol{\Sigma}_{X\mid Y}^{-1} \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1}$
*   $\boldsymbol{\Lambda}_{YX} = \boldsymbol{\Lambda}_{XY}^T$

Now, let's expand the exponent term: $-\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu})$.
Let $\tilde{\mathbf{x}} = \mathbf{x} - \boldsymbol{\mu}_X$ and $\tilde{\mathbf{y}} = \mathbf{y} - \boldsymbol{\mu}_Y$.

$$-\frac{1}{2} \begin{bmatrix} \tilde{\mathbf{x}} \\ \tilde{\mathbf{y}} \end{bmatrix}^T \begin{bmatrix} \boldsymbol{\Lambda}_{XX} & \boldsymbol{\Lambda}_{XY} \\ \boldsymbol{\Lambda}_{YX} & \boldsymbol{\Lambda}_{YY} \end{bmatrix} \begin{bmatrix} \tilde{\mathbf{x}} \\ \tilde{\mathbf{y}} \end{bmatrix}$$

$$= -\frac{1}{2} (\tilde{\mathbf{x}}^T \boldsymbol{\Lambda}_{XX} \tilde{\mathbf{x}} + \tilde{\mathbf{x}}^T \boldsymbol{\Lambda}_{XY} \tilde{\mathbf{y}} + \tilde{\mathbf{y}}^T \boldsymbol{\Lambda}_{YX} \tilde{\mathbf{x}} + \tilde{\mathbf{y}}^T \boldsymbol{\Lambda}_{YY} \tilde{\mathbf{y}})$$

Since we are interested in $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, the terms involving only $\mathbf{y}$ (i.e., $\tilde{\mathbf{y}}^T \boldsymbol{\Lambda}_{YY} \tilde{\mathbf{y}}$ and the $\mathbf{y}$ part of the cross terms) will be part of the normalization constant for the conditional distribution. We need to complete the square with respect to $\mathbf{x}$.

Let's rearrange the terms involving $\tilde{\mathbf{x}}$:

$$-\frac{1}{2} [\tilde{\mathbf{x}}^T \boldsymbol{\Lambda}_{XX} \tilde{\mathbf{x}} + \tilde{\mathbf{x}}^T \boldsymbol{\Lambda}_{XY} \tilde{\mathbf{y}} + (\tilde{\mathbf{x}}^T \boldsymbol{\Lambda}_{XY} \tilde{\mathbf{y}})^T]$$

$$= -\frac{1}{2} [\tilde{\mathbf{x}}^T \boldsymbol{\Lambda}_{XX} \tilde{\mathbf{x}} + 2\tilde{\mathbf{x}}^T \boldsymbol{\Lambda}_{XY} \tilde{\mathbf{y}}]$$

Now, recall the formula for completing the square: $(\mathbf{v} - \mathbf{a})^T \mathbf{M} (\mathbf{v} - \mathbf{a}) = \mathbf{v}^T \mathbf{M} \mathbf{v} - 2\mathbf{v}^T \mathbf{M} \mathbf{a} + \mathbf{a}^T \mathbf{M} \mathbf{a}$.

Comparing this with our expression, we have $\mathbf{v} = \tilde{\mathbf{x}}$, $\mathbf{M} = \boldsymbol{\Lambda}_{XX}$, and we want to find $\mathbf{a}$ such that $\mathbf{M} \mathbf{a} = -\boldsymbol{\Lambda}_{XY} \tilde{\mathbf{y}}$.

So, $\mathbf{a} = -\boldsymbol{\Lambda}_{XX}^{-1} \boldsymbol{\Lambda}_{XY} \tilde{\mathbf{y}}$.

Substituting the definitions of $\boldsymbol{\Lambda}_{XX}$ and $\boldsymbol{\Lambda}_{XY}$:

$\boldsymbol{\Lambda}_{XX}^{-1} = \boldsymbol{\Sigma}_{X\mid Y}$
$\boldsymbol{\Lambda}_{XY} = -\boldsymbol{\Sigma}_{X\mid Y}^{-1} \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1}$

Therefore,
$\mathbf{a} = - (\boldsymbol{\Sigma}_{X\mid Y}) (- \boldsymbol{\Sigma}_{X\mid Y}^{-1} \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1}) \tilde{\mathbf{y}}$
$\mathbf{a} = \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} \tilde{\mathbf{y}}$

So, the term in the exponent can be rewritten as:

$$-\frac{1}{2} (\tilde{\mathbf{x}} - \mathbf{a})^T \boldsymbol{\Lambda}_{XX} (\tilde{\mathbf{x}} - \mathbf{a}) + \text{terms independent of } \mathbf{x}$$

Substituting $\tilde{\mathbf{x}} = \mathbf{x} - \boldsymbol{\mu}_X$ and $\mathbf{a} = \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y)$:

$$-\frac{1}{2} ((\mathbf{x} - \boldsymbol{\mu}_X) - \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y))^T \boldsymbol{\Lambda}_{XX} ((\mathbf{x} - \boldsymbol{\mu}_X) - \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y))$$

This expression matches the quadratic form of a multivariate normal distribution if we define:

*   **Conditional Mean** $\boldsymbol{\mu}^*$:
    $$\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y)$$

*   **Conditional Covariance Matrix** $\boldsymbol{\Sigma}_{X\mid Y}$:
    $$\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Lambda}_{XX}^{-1} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$$

### Conclusion

Therefore, the conditional distribution $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$ is indeed a multivariate normal distribution with:

*   **Conditional Mean:**
    $$\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$$

*   **Conditional Covariance Matrix:**
    $$\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$$

And its probability density function is:

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X\mid Y}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$
