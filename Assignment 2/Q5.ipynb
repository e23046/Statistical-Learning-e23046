# 1. Support of (X)

The support of the triangular distribution is

$$
4 \le X \le 10
$$

Hence,

$$
{X=[4,10]}
$$

---

# 2. Probability Density Function

For a triangular distribution with parameters (a=4), (b=10), and (c=6),

$$
f_X(x)=
\begin{cases}
\dfrac{2(x-4)}{(10-4)(6-4)}, & 4\le x\le6,\\[8pt]
\dfrac{2(10-x)}{(10-4)(10-6)}, & 6<x\le10,\\[8pt]
0, & \text{otherwise}.
\end{cases}
$$

Simplifying,

$$
f_X(x)=
\begin{cases}
\dfrac{x-4}{6}, & 4\le x\le6,\\[8pt]
\dfrac{10-x}{12}, & 6<x\le10,\\[8pt]
0, & \text{otherwise}.
\end{cases}
$$

---

# 3. Verification that (f_X) Integrates to 1

$$
\int_4^{10} f_X(x)\,dx
=
\int_4^6 \frac{x-4}{6}\,dx
+
\int_6^{10}\frac{10-x}{12}\,dx
$$

First integral:

$$
\int_4^6 \frac{x-4}{6}\,dx
=
\frac{1}{6}\left[\frac{(x-4)^2}{2}\right]_4^6
=
\frac{1}{6}\times\frac{(6-4)^2}{2}
=
\frac{1}{6}\times\frac{4}{2}
=
\frac13
$$

Second integral:

$$
\int_6^{10}\frac{10-x}{12}\,dx
=
\frac{1}{12}\left[10x-\frac{x^2}{2}\right]_6^{10}
=
\frac{1}{12}\left(\left(100-\frac{100}{2}\right)-\left(60-\frac{36}{2}\right)\right)
=
\frac{1}{12}\left(50-42\right)
=
\frac{8}{12}
=
\frac23
$$

Thus,

$$
\frac13+\frac23=1
$$

Hence verified.

# 4. Expectation (E[X])

Substitute values:

$$

$${E[X]=\frac{20}{3}\approx6.67\text{ hours}}

# 5. Variance ($\mathrm{Var}(X)$)

For a triangular distribution,

$$
\mathrm{Var}(X)=
\frac{a^2+b^2+c^2-ab-ac-bc}{18}
$$

Substitute values:

$$
\frac{4^2+10^2+6^2-(4)(10)-(4)(6)-(10)(6)}{18}
$$

$$
=\frac{16+100+36-40-24-60}{18}

$$

$$
{\mathrm{Var}(X)=\frac{14}{9}\approx1.56}

# 8. Probability that Installation Takes at Most 5 Hours

$$
P(X\le5)
=
\int_4^5 \frac{x-4}{6}\,dx
$$

$$
=\frac{1}{6}\left[\frac{(x-4)^2}{2}\right]_4^5
$$

$$
=\frac{1}{6}\times\frac{(5-4)^2}{2}
=\frac{1}{6}\times\frac12
=\frac{1}{12}
$$

$$
{P(X\le5)=\frac{1}{12}\approx0.0833}
$$

---

# 9. Probability that Installation Takes Between 5 and 8 Hours

$$
P(5\le X\le8)
=
\int_5^6 \frac{x-4}{6}\,dx
+
\int_6^8 \frac{10-x}{12}\,dx
$$

First part:

$$
\int_5^6 \frac{x-4}{6}\,dx
=\frac{1}{6}\left[\frac{(x-4)^2}{2}\right]_5^6
=\frac{1}{6}\left(\frac{(6-4)^2}{2}-\frac{(5-4)^2}{2}\right)
=\frac{1}{6}\left(\frac{4}{2}-\frac{1}{2}\right)
=\frac{1}{6}\times\frac{3}{2}
=\frac{1}{4}
$$

Second part:

$$
\int_6^8 \frac{10-x}{12}\,dx
=\frac{1}{12}\left[10x-\frac{x^2}{2}\right]_6^8
=\frac{1}{12}\left(\left(80-\frac{64}{2}\right)-\left(60-\frac{36}{2}\right)\right)
=\frac{1}{12}\left((80-32)-(60-18)\right)
=\frac{1}{12}\left(48-42\right)
=\frac{6}{12}
=\frac12
$$

Therefore,

$$
P(5\le X\le8)
=\frac14+\frac12
=\frac34
$$

$$
{P(5\le X\le8)=0.75}
$$

# 10. Use in Project Planning

This model helps to:

## Plotting the Triangular Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the triangular distribution
a = 4
b = 10
c = 6

def triangular_pdf(x, a, b, c):
    """Probability Density Function (PDF) of a triangular distribution."""
    if a <= x <= c:
        return 2 * (x - a) / ((b - a) * (c - a))
    elif c < x <= b:
        return 2 * (b - x) / ((b - a) * (b - c))
    else:
        return 0

# Generate x values for the plot
x_values = np.linspace(a - 1, b + 1, 500) # Extend range slightly to show zero probability

# Calculate y values (PDF) for each x
y_values = [triangular_pdf(x, a, b, c) for x in x_values]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=f'Triangular PDF (a={a}, b={b}, c={c})', color='blue')

# Add markers for a, b, c
plt.axvline(a, color='red', linestyle='--', label=f'a = {a} (Min)')
plt.axvline(b, color='green', linestyle='--', label=f'b = {b} (Max)')
plt.axvline(c, color='purple', linestyle='--', label=f'c = {c} (Mode)')

plt.title('Probability Density Function of a Triangular Distribution')
plt.xlabel('X (Installation Time in Hours)')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.ylim(bottom=0) # Ensure y-axis starts at 0
plt.xlim(a-1, b+1)
plt.show()
```
