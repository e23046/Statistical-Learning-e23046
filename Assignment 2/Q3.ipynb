# 1. Construction of the Probability Space

Let the sample space be

$$ \Omega={0,1,2,3} $$

where each outcome represents the number of emergency transformer failures reported in one day.

Let the sigma-algebra be

$$ \mathfrak F = 2^\Omega $$

(the power set of $(\Omega)$), containing all subsets of $(\Omega)$.

Define the probability measure (P) by

$$ P(0)=0.50,\qquad P(1)=0.30,\qquad P(2)=0.15,\qquad P(3)=0.05 $$

Check:

$$ 0.50+0.30+0.15+0.05=1 $$

Hence,

$$ (\Omega,\mathfrak F,P) $$

is a valid probability space.

---

# 2. Definition of the Random Variable

Define the random variable

$$ X:\Omega \to \mathbb R $$

by

$$ X(\omega)=\omega,\qquad \omega\in\Omega $$

Thus, (X) gives the number of transformer failures in a day.

---

# 3. Probability Mass Function (PMF)

The probability mass function of (X) is

$$ p_X(x)=P(X=x) $$

Hence,

$$ p_X(x)=\begin{cases}0.50, & x=0\_\text{4pt}\\0.30, & x=1\_\text{4pt}\\0.15, & x=2\_\text{4pt}\\0.05, & x=3\_\text{4pt}\\0, & \text{otherwise}\end{cases} $$

---

# 4. Expectation (E[X])

The expectation of a discrete random variable is

$$ E[X]=\sum xP(X=x) $$

Therefore,

$$ E[X]
=0(0.50)+1(0.30)+2(0.15)+3(0.05) $$

$$ =0+0.30+0.30+0.15 $$

$$ E[X]=0.75 $$

---

# 5. Variance ($\mathrm{Var}(X)$)

First compute ($E[X^2]$):

$$ E[X^2]=\sum x^2P(X=x) $$

$$ =0^2(0.50)+1^2(0.30)+2^2(0.15)+3^2(0.05) $$

$$ =0+0.30+0.60+0.45 $$

$$ E[X^2]=1.35 $$

Now,

$$ \mathrm{Var}(X)=E[X^2]-(E[X])^2 $$

$$ =1.35-(0.75)^2 $$

$$ =1.35-0.5625 $$

$$ \boxed{\mathrm{Var}(X)=0.7875} $$

---

# 6. Required Probabilities

## (i) Probability that at least one failure occurs

$$ P(X\ge1)=P(X=1)+P(X=2)+P(X=3) $$

$$ =0.30+0.15+0.05 $$

$$ \boxed{P(X\ge1)=0.50} $$

---

## (ii) Probability that at least two failures occur

$$ P(X\ge2)=P(X=2)+P(X=3) $$

$$ =0.15+0.05 $$

$$ \boxed{P(X\ge2)=0.20} $$

---

# 7. Practical Meaning of Expectation and Variance

## Expectation

The expectation

$$ E[X]=0.75 $$

means that, on average, the electricity distribution district experiences about **0.75 transformer failures per day**.

---

## Variance

The variance

$$ \mathrm{Var}(X)=0.7875 $$

measures the variability of daily transformer failures around the average value. A higher variance indicates that the number of failures changes significantly from day to day.
