Answer For The question 07

Given:

* Minimum time: \(a = 5\)
* Most likely time: \(c = 11\)
* Expected time condition: \(E[X] = c\)

---

## **1. Find maximum time (b)**

PERT mean formula:

$$
E[X] = \frac{a + 4c + b}{6}
$$

Set \(E[X] = c\):

$$
c = \frac{a + 4c + b}{6}
$$

Multiply by 6:

$$
6c = a + 4c + b
$$

Rearrange:

$$
2c = a + b
$$

$$
b = 2c - a
$$

Substitute values:

$$
b = 2(11) - 5 = 22 - 5 = 17
$$

### ✔ **Maximum time:**

$$
b = 17 \text{ days}
$$

---

## **2. Variance of the distribution**

Since \(c\) becomes exactly the midpoint:
$$
c = \frac{a+b}{2} = \frac{5+17}{2} = 11
$$
the distribution is **symmetric**.

Using the Beta-based PERT variance form:

$$
\mathrm{Var}(X) = \frac{(c - a)(b - c)}{7}
$$

Substitute:

$$
\mathrm{Var}(X)=\frac{(11-5)(17-11)}{7}
$$

$$
=\frac{6 \cdot 6}{7}
=\frac{36}{7}
\approx 5.143
$$

---

## ✔ **Final Answers**

### **(1) Maximum time**

$$
{b = 17 \text{ days}}
$$

### **(2) Variance**

$$
{\mathrm{Var}(X)=\frac{36}{7} \approx 5.143 \text{ days}^2}
$$
