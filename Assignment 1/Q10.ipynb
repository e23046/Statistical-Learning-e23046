## Answers to Theoretical Questions

### 1. Probability Space (Ω, F, P) Explicitly

The probability space is given as $(\Omega, \mathcal{F}, \mathbb{P})$ where:

*   $\Omega = \mathbb{R}^2$ (the sample space, representing all possible pairs of temperature and pressure deviations).
*   $\mathcal{F} = \mathcal{B}(\mathbb{R}^2)$ (the sigma-algebra of Borel sets on $\mathbb{R}^2$, which allows us to measure probabilities of events).
*   $\mathbb{P}$ is the probability measure induced by the joint probability density function $f_{T,P}(t,p)$:

    $$f_{T,P}(t,p) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right), \quad (t,p) \in \mathbb{R}^2$$

    For any event $A \in \mathcal{F}$, the probability is given by:

    $$ \mathbb{P}(A) = \iint_A f_{T,P}(t,p) \, dt \, dp $$

### 2. Compute the σ-algebra generated by X, namely σ(X)

The random variable $X(t,p) = t$ records only the reactor temperature deviation. The $\sigma$-algebra generated by $X$, denoted $\sigma(X)$, consists of all sets of the form $X^{-1}(B)$ where $B \in \mathcal{B}(\mathbb{R})$ (the Borel sets on $\mathbb{R}$).

Since $X(t,p) = t$, a set $B \in \mathcal{B}(\mathbb{R})$ corresponds to $X^{-1}(B) = \{(t,p) \in \mathbb{R}^2 \mid t \in B \} = B \times \mathbb{R}$.

Therefore, $\sigma(X) = \{ B \times \mathbb{R} \mid B \in \mathcal{B}(\mathbb{R}) \}$. These are "vertical strips" in the $(t,p)$-plane.

### 3. Compute the σ-algebra generated by Y, namely σ(Y)

The random variable $Y(t,p) = (t,p)$ records both temperature and pressure deviations. The $\sigma$-algebra generated by $Y$, denoted $\sigma(Y)$, consists of all sets of the form $Y^{-1}(C)$ where $C \in \mathcal{B}(\mathbb{R}^2)$.

Since $Y(t,p) = (t,p)$, $Y^{-1}(C) = C$. Therefore, $\sigma(Y) = \mathcal{B}(\mathbb{R}^2)$. This means $\sigma(Y)$ is the entire Borel $\sigma$-algebra on $\mathbb{R}^2$, which is the original $\sigma$-algebra $\mathcal{F}$ of the probability space.

### 4. Show that σ(X) ⊆ σ(Y), and explain why Y contains more information than X

#### Showing σ(X) ⊆ σ(Y)

From the previous answers:
*   $\sigma(X) = \{ B \times \mathbb{R} \mid B \in \mathcal{B}(\mathbb{R}) \}$
*   $\sigma(Y) = \mathcal{B}(\mathbb{R}^2)$

Any set of the form $B \times \mathbb{R}$ where $B \in \mathcal{B}(\mathbb{R})$ is a Borel set in $\mathbb{R}^2$. This is because $\mathcal{B}(\mathbb{R}^2)$ is the smallest $\sigma$-algebra containing all open rectangles (or products of Borel sets from $\mathbb{R}$). Specifically, if $B$ is a Borel set in $\mathbb{R}$ and $\mathbb{R}$ is a Borel set in $\mathbb{R}$, then their product $B \times \mathbb{R}$ is a Borel set in $\mathbb{R}^2$.

Thus, every set in $\sigma(X)$ is also a set in $\sigma(Y)$. Therefore, $\sigma(X) \subseteq \sigma(Y)$.

#### Explaining why Y contains more information than X

The $\sigma$-algebra generated by a random variable represents the information contained in that random variable. A coarser $\sigma$-algebra means less information, while a finer (larger) $\sigma$-algebra means more information.

*   $\sigma(X)$ only allows us to distinguish events based *only* on the temperature deviation $t$. For example, we can know if $t$ is in a certain range, but we cannot distinguish between states where $t$ is the same but $p$ is different.
*   $\sigma(Y)$ allows us to distinguish *any* event that can be defined on $\mathbb{R}^2$ concerning both $t$ and $p$. We can know the exact pair $(t,p)$.

Since $\sigma(X)$ is a proper subset of $\sigma(Y)$, it means that $Y$ allows us to distinguish between more events (it provides a finer partition of the sample space) than $X$. In other words, knowing the full sensor state $(T,P)$ (represented by $Y$) provides all the information that knowing only $T$ (represented by $X$) would, plus additional information about $P$. Therefore, $Y$ contains more information than $X$ because it captures both dimensions of the sensor state, whereas $X$ only captures one.

### 5. Find the marginal density of X

$f_T(t) = \frac{1}{\sqrt{2\pi \cdot 1^2}} \exp\left(-\frac{t^2}{2 \cdot 1^2}\right) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{t^2}{2}\right)$

$f_P(p) = \frac{1}{\sqrt{2\pi \cdot 4^2}} \exp\left(-\frac{p^2}{2 \cdot 4^2}\right) = \frac{1}{\sqrt{32\pi}} \exp\left(-\frac{p^2}{32}\right)$.

If $P \sim N(0, \sigma^2)$ where $\sigma^2 = 4$, then $\sigma=2$.

$f_P(p) = \frac{1}{\sigma\sqrt{2\pi}} \exp\left(-\frac{p^2}{2\sigma^2}\right) = \frac{1}{2\sqrt{2\pi}} \exp\left(-\frac{p^2}{2 \cdot 4}\right) = \frac{1}{2\sqrt{2\pi}} \exp\left(-\frac{p^2}{8}\right)$

Now, let's check the product:

$f_T(t)f_P(p) = \left(\frac{1}{\sqrt{2\pi}} \exp\left(-\frac{t^2}{2}\right)\right) \left(\frac{1}{2\sqrt{2\pi}} \exp\left(-\frac{p^2}{8}\right)\right) = \frac{1}{2 \cdot 2\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right)$.

This matches the given $f_{T,P}(t,p)$.

Since $X(t,p) = t$, the random variable $X$ is identical to $T$. Therefore, the marginal density of $X$ is the marginal density of $T$:

This is the probability density function of a standard normal distribution, $N(0,1)$.

Alternatively, we can find the marginal density by integrating the joint density over $p$:

$f_X(x) = \int_{-\infty}^{\infty} f_{T,P}(x,p) \, dp = \int_{-\infty}^{\infty} \frac{1}{4\pi} \exp\left(-\frac{x^2}{2} - \frac{p^2}{8}\right) \, dp$
$f_X(x) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{x^2}{2}\right) \int_{-\infty}^{\infty} \frac{1}{2\sqrt{2\pi}} \exp\left(-\frac{p^2}{8}\right) \, dp$

The integral part is the PDF of $N(0,4)$ (variance 4, std dev 2) integrated over its entire domain, which equals 1.

So, $f_X(x) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{x^2}{2}\right)$.



### 6. Find the marginal density of Y

The random variable $Y(t,p) = (t,p)$ is simply the pair of underlying random variables $(T,P)$. Thus, the marginal density of $Y$ is its joint density, which is precisely the given joint density $f_{T,P}(t,p)$.

$$ f_Y(t,p) = f_{T,P}(t,p) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right) \quad \text{for } (t,p) \in \mathbb{R}^2 $$

### 7. Compute the following probabilities:

We use the fact that $X \sim N(0,1)$ and $P \sim N(0,4)$ (which means $P \sim N(0, 2^2)$).

#### P(X ≤ 0)
Since $X \sim N(0,1)$, its distribution is symmetric about 0.

$P(X \leq 0) = \Phi(0) = 0.5$, where $\Phi$ is the CDF of the standard normal distribution.

#### P(X > 1)
$P(X > 1) = 1 - P(X \leq 1) = 1 - \Phi(1)$.

Using a standard normal table or calculator, $\Phi(1) \approx 0.8413$.

$P(X > 1) \approx 1 - 0.8413 = 0.1587$.

#### P(Y ∈ (−∞,0] × R)
$P(Y \in (-\infty,0] \times \mathbb{R})$ means $P(T \leq 0 \text{ and } P \in \mathbb{R})$.
Since $T$ and $P$ are independent, and $P$ always takes values in $\mathbb{R}$ with probability 1:

$P(Y \in (-\infty,0] \times \mathbb{R}) = P(T \leq 0) \cdot P(P \in \mathbb{R}) = P(T \leq 0) \cdot 1$.

Since $T \sim N(0,1)$, $P(T \leq 0) = 0.5$.

So, $P(Y \in (-\infty,0] \times \mathbb{R}) = 0.5$.

#### P(Y ∈ [−1,1] × [−2,2])
$P(Y \in [-1,1] \times [-2,2])$ means $P(-1 \leq T \leq 1 \text{ and } -2 \leq P \leq 2)$.
Due to independence:

$P(Y \in [-1,1] \times [-2,2]) = P(-1 \leq T \leq 1) \cdot P(-2 \leq P \leq 2)$.

For $T \sim N(0,1)$:
$P(-1 \leq T \leq 1) = \Phi(1) - \Phi(-1) = \Phi(1) - (1 - \Phi(1)) = 2\Phi(1) - 1 \approx 2(0.8413) - 1 = 1.6826 - 1 = 0.6826$.

For $P \sim N(0,4)$, i.e., $N(0, 2^2)$. We need to standardize $P$ to use the standard normal CDF.
Let $Z = P/2$, then $Z \sim N(0,1)$.
$P(-2 \leq P \leq 2) = P(-2/2 \leq P/2 \leq 2/2) = P(-1 \leq Z \leq 1)$.
This is the same probability as $P(-1 \leq T \leq 1)$ for $T \sim N(0,1)$.

So, $P(-2 \leq P \leq 2) \approx 0.6826$.

Therefore,
$P(Y \in [-1,1] \times [-2,2]) \approx 0.6826 \cdot 0.6826 \approx 0.4660$.

These probabilities are related to the empirical rule (68-95-99.7 rule) for normal distributions, where approximately 68% of the data falls within one standard deviation of the mean.

### 8. Compute the conditional distribution of X given Y=(t,p)

The random variable $X(t,p) = t$. The random variable $Y(t,p) = (t,p)$.

If we are given that $Y=(t_0, p_0)$ for some specific values $t_0, p_0$, it means we know the exact values of both temperature and pressure deviations.

Since $X$ is defined as the temperature deviation, i.e., $X=T$, then if $Y=(t_0, p_0)$ is known, we know $T=t_0$. Therefore, $X=t_0$ is known with certainty.

The conditional distribution of $X$ given $Y=(t_0,p_0)$ is a degenerate distribution (a Dirac delta function) at $t_0$.

$$ P(X = x \mid Y=(t_0, p_0)) = \begin{cases} 1 & \text{if } x = t_0 \\ 0 & \text{if } x \neq t_0 \end{cases} $$

Its probability density function would be $f_{X|Y}(x \mid t_0, p_0) = \delta(x - t_0)$, where $\delta$ is the Dirac delta function. In simpler terms, if you know $Y$, you know $X$ precisely.

### 9. Compute the conditional distribution of Y given X=t

We want to find the conditional distribution of $Y=(T,P)$ given $X=t_0$, where $X=T$. This is equivalent to finding the conditional distribution of $(T,P)$ given $T=t_0$.

Using the formula for conditional density for continuous variables:

$f_{Y|X}(t', p \mid t_0) = f_{T,P|T}(t', p \mid t_0) = \frac{f_{T,P}(t',p)}{f_T(t')} \quad \text{evaluated at } t'=t_0$

So, $f_{Y|X}(t_0, p \mid t_0) = \frac{f_{T,P}(t_0,p)}{f_T(t_0)}$.

We know:
$f_{T,P}(t_0,p) = \frac{1}{4\pi} \exp\left(-\frac{t_0^2}{2} - \frac{p^2}{8}\right) = \left(\frac{1}{\sqrt{2\pi}} \exp\left(-\frac{t_0^2}{2}\right)\right) \left(\frac{1}{2\sqrt{2\pi}} \exp\left(-\frac{p^2}{8}\right)\right) = f_T(t_0) f_P(p)$.

Substitute this into the conditional density formula:

$f_{Y|X}(t_0, p \mid t_0) = \frac{f_T(t_0) f_P(p)}{f_T(t_0)} = f_P(p)$.

This means that given $X=t_0$ (i.e., $T=t_0$), the conditional distribution of $Y=(T,P)$ is such that $T$ is fixed at $t_0$, and $P$ follows its original marginal distribution $N(0,4)$. This makes sense because $T$ and $P$ are independent. Knowing $T$ tells you nothing new about $P$. So, the conditional distribution of $P$ given $T=t_0$ is simply the marginal distribution of $P$.

Thus, the conditional distribution of $Y$ given $X=t_0$ is a distribution where the first component is fixed at $t_0$, and the second component $P$ follows $N(0,4)$.

$$ f_{Y|X}((t',p) \mid t_0) = \delta(t' - t_0) \cdot f_P(p) = \delta(t' - t_0) \cdot \frac{1}{2\sqrt{2\pi}} \exp\left(-\frac{p^2}{8}\right) $$

This again involves a Dirac delta function for the $T$ component, signifying that $T$ is precisely known to be $t_0$, while the $P$ component remains a continuous random variable distributed as $N(0,4)$. This conditional distribution is effectively $P \sim N(0,4)$ when $T=t_0$ is known.

### 10. Explain what is known when one conditions on X and what additional information becomes available when one conditions on Y

#### When one conditions on X (i.e., $X=t_0$):
When we condition on $X=t_0$, we know the exact value of the reactor temperature deviation. That is, we know $T=t_0$. Since $T$ and $P$ are independent, knowing $T$ does **not** provide any information about $P$. Therefore, the probability distribution of $P$ remains unchanged, i.e., $P \sim N(0,4)$. In summary, conditioning on $X=t_0$ tells us *only* the temperature deviation; the pressure deviation remains a random variable following its original distribution.

#### When one conditions on Y (i.e., $Y=(t_0,p_0)$):
When we condition on $Y=(t_0,p_0)$, we know the exact values of *both* the reactor temperature deviation and the reactor pressure deviation. That is, we know $T=t_0$ and $P=p_0$. In this case, there is no remaining uncertainty about either component of the sensor state. Both $T$ and $P$ are fixed at specific values. This provides the maximum possible information about the system's state as measured by these two sensors.

### 11. Show that $X = \pi_1 \circ Y$, where $\pi_1(t,p)=t$, and explain why this implies that X is a measurable function of Y

#### Showing $X = \pi_1 \circ Y$
Let's define the functions:
*   $X: \Omega \to \mathbb{R}$ by $X(t,p) = t$
*   $Y: \Omega \to \mathbb{R}^2$ by $Y(t,p) = (t,p)$
*   $\pi_1: \mathbb{R}^2 \to \mathbb{R}$ by $\pi_1(u,v) = u$ (This is the projection map onto the first coordinate).

Now, let's compute the composition $\pi_1 \circ Y$:

$(\pi_1 \circ Y)(t,p) = \pi_1(Y(t,p)) = \pi_1((t,p)) = t$.

Since $X(t,p) = t$, we have shown that $X = \pi_1 \circ Y$.

#### Explaining why this implies that X is a measurable function of Y

In probability theory, a random variable $X$ is said to be a measurable function of another random variable $Y$ if $\sigma(X) \subseteq \sigma(Y)$. This means that all information contained in $X$ is also contained in $Y$. If $X = g(Y)$ for some measurable function $g$, then $X$ is measurable with respect to $\sigma(Y)$.

In our case, $X = \pi_1 \circ Y$, where $\pi_1$ is the projection map. The projection map $\pi_1: \mathbb{R}^2 \to \mathbb{R}$ is a continuous function, and any continuous function is measurable. Specifically, for any Borel set $B \in \mathcal{B}(\mathbb{R})$, the pre-image $\pi_1^{-1}(B) = B \times \mathbb{R}$, which is a Borel set in $\mathbb{R}^2$. Thus, $\pi_1$ is a $\mathcal{B}(\mathbb{R}^2)/\mathcal{B}(\mathbb{R})$-measurable function.

Since $Y$ is a $\mathcal{F}/\mathcal{B}(\mathbb{R}^2)$-measurable function (it's the identity map, so it generates $\mathcal{F}$), and $\pi_1$ is a $\mathcal{B}(\mathbb{R}^2)/\mathcal{B}(\mathbb{R})$-measurable function, their composition $X = \pi_1 \circ Y$ is also a $\mathcal{F}/\mathcal{B}(\mathbb{R})$-measurable function.

This composition implies that any event detectable by $X$ is formed by taking a set $B \in \mathcal{B}(\mathbb{R})$, finding its pre-image under $\pi_1$ (which is $B \times \mathbb{R}$), and then taking the pre-image of that under $Y$ (which is just $B \times \mathbb{R}$ itself). Since $B \times \mathbb{R} \in \mathcal{B}(\mathbb{R}^2) = \sigma(Y)$, it means that every event in $\sigma(X)$ is also in $\sigma(Y)$.

Therefore, $X$ is a measurable function of $Y$, which formally proves that $X$ provides no information that is not already present in $Y$. All information in $X$ can be derived from $Y$ by simply taking its first component.

## Python-based Visualization

Let's visualize the joint density of Y=(T,P), the marginal density of X, and illustrate the sigma-algebras.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import matplotlib.patches as patches
import matplotlib.lines as mlines

# Define the joint PDF f_T,P(t,p)
def joint_pdf(t, p):
    # T ~ N(0,1), P ~ N(0,4) (std dev = 2)
    # f_T(t) = 1/sqrt(2*pi) * exp(-t^2/2)
    # f_P(p) = 1/(2*sqrt(2*pi)) * exp(-p^2/8)
    return (1 / (4 * np.pi)) * np.exp(-t**2 / 2 - p**2 / 8)

# Define the marginal PDF f_X(x) (which is f_T(t))
def marginal_pdf_X(x):
    return norm.pdf(x, loc=0, scale=1)

# 1. Joint density of Y=(T,P) as a contour plot

# Create a grid for T and P values
t_vals = np.linspace(-4, 4, 100)
p_vals = np.linspace(-8, 8, 100) # 2 standard deviations for P is 4, 3 is 6, going a bit wider
T_grid, P_grid = np.meshgrid(t_vals, p_vals)

# Calculate joint PDF values on the grid
Z = joint_pdf(T_grid, P_grid)

plt.figure(figsize=(10, 8))
contour_plot = plt.contourf(T_grid, P_grid, Z, levels=20, cmap='viridis')
plt.colorbar(contour_plot, label='Joint Probability Density $f_{T,P}(t,p)$')
plt.xlabel('Temperature Deviation (t)')
plt.ylabel('Pressure Deviation (p)')
plt.title('Contour Plot of Joint Density of Y=(T,P)')
plt.grid(True, linestyle=':', alpha=0.6)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(t_vals, marginal_pdf_X(t_vals), color='blue', linewidth=2)
plt.xlabel('Temperature Deviation (x)')
plt.ylabel('Probability Density $f_X(x)$')
plt.title('Marginal Density of X (Temperature Deviation)')
plt.grid(True, linestyle=':', alpha=0.6)
plt.fill_between(t_vals, 0, marginal_pdf_X(t_vals), alpha=0.3, color='lightblue')
plt.show()
```

## Differential Entropy Calculations

Differential entropy is the analogue of discrete entropy for continuous random variables. For a continuous random variable $X$ with PDF $f_X(x)$, the differential entropy is defined as:

$h(X) = - \int_{\mathbb{R}} f_X(x) \ln(f_X(x)) \, dx$

And for a multivariate random variable $Y$ with PDF $f_Y(y)$, it's:

$h(Y) = - \int_{\mathbb{R}^n} f_Y(y) \ln(f_Y(y)) \, dy$

For a normal distribution $N(\mu, \sigma^2)$, the differential entropy is known to be:

$h(N(\mu, \sigma^2)) = \frac{1}{2} \ln(2\pi e \sigma^2)$.

### 1. Compute the differential entropy of X: $h(X)$

$X \sim N(0,1)$, so $\mu = 0$ and $\sigma^2 = 1$.

$h(X) = \frac{1}{2} \ln(2\pi e \cdot 1) = \frac{1}{2} \ln(2\pi e)$.

### 2. Compute the differential entropy of Y: $h(Y)$

$Y=(T,P)$, where $T \sim N(0,1)$ and $P \sim N(0,4)$ are independent. The joint distribution is a multivariate normal distribution with mean vector $\mathbf{\mu} = (0,0)^T$ and covariance matrix $\mathbf{\Sigma} = \begin{pmatrix} 1 & 0 \\ 0 & 4 \end{pmatrix}$.

The differential entropy of a multivariate normal distribution $N(\mathbf{\mu}, \mathbf{\Sigma})$ is given by:

$h(Y) = \frac{1}{2} \ln((2\pi e)^k \det(\mathbf{\Sigma}))$, where $k$ is the dimension.

Here, $k=2$. The determinant of the covariance matrix is $\det(\mathbf{\Sigma}) = 1 \cdot 4 - 0 \cdot 0 = 4$.

So,

$h(Y) = \frac{1}{2} \ln((2\pi e)^2 \cdot 4) = \frac{1}{2} \ln(4 \cdot (2\pi e)^2)$.

Using properties of logarithms:

$h(Y) = \frac{1}{2} [\ln(4) + \ln((2\pi e)^2)] = \frac{1}{2} [\ln(2^2) + 2 \ln(2\pi e)]$

$h(Y) = \frac{1}{2} [2\ln(2) + 2 \ln(2\pi e)] = \ln(2) + \ln(2\pi e)$.

Alternatively, for independent random variables, the joint entropy is the sum of their individual entropies:

$h(Y) = h(T) + h(P)$.

$h(T) = h(N(0,1)) = \frac{1}{2} \ln(2\pi e \cdot 1) = \frac{1}{2} \ln(2\pi e)$.

$h(P) = h(N(0,4)) = \frac{1}{2} \ln(2\pi e \cdot 4) = \frac{1}{2} [\ln(2\pi e) + \ln(4)] = \frac{1}{2} \ln(2\pi e) + \ln(2)$.

$h(Y) = \frac{1}{2} \ln(2\pi e) + \frac{1}{2} \ln(2\pi e) + \ln(2) = \ln(2\pi e) + \ln(2)$.

Both methods yield the same result.

### 3. Compute the conditional differential entropy $h(Y|X)$, and interpret it physically.

The conditional differential entropy $h(Y|X)$ is defined as:

$h(Y|X) = \mathbb{E}[-\ln(f_{Y|X}(Y|X))] = - \int_{\mathbb{R}} \int_{\mathbb{R}} f_{X,Y}(x,y) \ln(f_{Y|X}(y|x)) \, dy \, dx$

We also know that $h(Y|X) = h(Y) - h(X)$ (for continuous random variables, analogous to $H(Y|X) = H(Y) - H(X)$ for discrete variables when the chain rule for entropy is applied).

$h(Y|X) = (\ln(2\pi e) + \ln(2)) - (\frac{1}{2} \ln(2\pi e)) = \frac{1}{2} \ln(2\pi e) + \ln(2)$.

Let's verify this using the conditional distribution we found earlier. Given $X=t_0$ (i.e., $T=t_0$), the conditional distribution of $Y=(T,P)$ is such that $T$ is fixed at $t_0$, and $P$ follows $N(0,4)$. In essence, $f_{Y|X}(t_0, p \mid t_0) = f_P(p)$.

So, $h(Y|X=t_0) = h(P) = \frac{1}{2} \ln(2\pi e \cdot 4)$.

Since this conditional entropy does not depend on $t_0$, the expectation over $X$ is just this value.

$h(Y|X) = h(P) = \frac{1}{2} \ln(2\pi e \cdot 4) = \frac{1}{2} \ln(2\pi e) + \ln(2)$. This matches.

#### Physical Interpretation of $h(Y|X)$

$h(Y|X)$ represents the remaining uncertainty about $Y$ (the full sensor state, $(T,P)$) after $X$ (the temperature deviation, $T$) has been observed. Since $T$ and $P$ are independent, knowing $T$ reduces the uncertainty about $Y$ *only* by fixing the value of $T$. The uncertainty associated with $P$ remains unchanged.

Therefore, $h(Y|X)$ is precisely the entropy of the pressure deviation $P$. It quantifies the amount of uncertainty that still exists in the system's state ($Y$) when only the temperature deviation ($X$) is known. This uncertainty is entirely due to the unknown pressure deviation ($P$). If $P$ were also known, this entropy would be zero.

### 4. Discuss what happens to $h(X|Y)$ in this model, and explain why the continuous case differs from the discrete case when one random variable is completely determined by the other.

#### What happens to $h(X|Y)$ in this model?

We know that $X = \pi_1(Y)$, which means $X$ is completely determined by $Y$. If $Y=(t_0, p_0)$ is known, then $X$ is precisely $t_0$. There is no uncertainty about $X$ once $Y$ is known.

Therefore, the conditional differential entropy $h(X|Y)$ should be 0.

Let's check this using the formula $h(X|Y) = h(X) + h(Y) - h(Y) = 0$ (This is not quite right for continuous variables, $I(X;Y)=h(X)-h(X|Y)=h(Y)-h(Y|X)$, and if $X=g(Y)$ then $I(X;Y)=h(X)$ so $h(X|Y)=0$).

More formally, if $X = g(Y)$, then the conditional density $f_{X|Y}(x|y)$ is a Dirac delta function centered at $g(y)$.

$f_{X|Y}(x|y) = \delta(x - g(y))$.

The differential entropy of a distribution with a Dirac delta function is generally taken to be $-\infty$ or is undefined in the standard definition, but it is often considered to be 0 in the context of information theory, meaning no uncertainty. A more rigorous approach involves sequences of distributions converging to a delta function.

So, $h(X|Y) = 0$.

#### Why the continuous case differs from the discrete case when one random variable is completely determined by the other.

In the discrete case, if $X=g(Y)$, then $H(X|Y) = 0$. This means that knowing $Y$ leaves no uncertainty about $X$. This is straightforward and intuitive.

In the continuous case, if $X=g(Y)$ (where $g$ is a smooth function), the conditional differential entropy $h(X|Y)$ is also 0. This implies that there is no remaining *information* or *uncertainty* about $X$ once $Y$ is known. However, this differs subtly from the discrete case in how we interpret the entropy itself.

Differential entropy has a few properties that make it different from discrete entropy:

1.  **It can be negative:** Unlike discrete entropy, differential entropy can be negative. This happens when the probability density is sharply concentrated, making $\ln(f(x))$ large and negative over a significant region. A degenerate distribution (like a Dirac delta function) can be seen as an extreme case of concentration.

2.  **It is not absolute but relative:** Differential entropy is sensitive to the choice of units. If you change the units of measurement for the random variable, the differential entropy changes. This is because the integral involves $\ln(f_X(x))$, and $f_X(x)$ has units of (1/unit of $x$). For example, if $X$ is measured in meters, $f_X(x)$ is in $1/m$, and $\ln(f_X(x))$ involves $\ln(1/m) = -\ln(m)$.

When $X$ is completely determined by $Y$, the conditional distribution of $X$ given $Y$ is a point mass (a Dirac delta function). The density function of a point mass is infinite at that point and zero everywhere else, and its integral is 1. The $\ln$ of an infinite density is infinite. The traditional integral for entropy $ -\int f \ln f $ for a Dirac delta function technically diverges to $-\infty$ if you apply it directly without careful limits. However, in the context of information theory, we consider this to represent *zero uncertainty* or *zero information* left to gain, so $h(X|Y)=0$.

The difference is mainly in the mathematical rigor of applying the integral definition to a degenerate distribution. While for discrete variables, a probability of 1 for a single outcome clearly leads to an entropy of 0, for continuous variables, a probability *density* that is infinite at a point requires special handling or reinterpretation to yield 0. The intuitive meaning, however, remains consistent: knowing $Y$ completely resolves $X$'s uncertainty.
