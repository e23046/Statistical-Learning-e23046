# Bertrand's Paradox

A chord of a circle is chosen “at random.” What is the probability that its length exceeds the side length of the inscribed equilateral triangle?

This problem, known as Bertrand's Paradox, highlights how the phrase "chosen at random" can lead to different results depending on the specific method used to define randomness. We will explore three common and mathematically natural methods for selecting a random chord and show how each yields a different probability.

## Geometric Setup

Let's define a circle with radius `R` centered at the origin (0,0). We will also consider an equilateral triangle inscribed within this circle. We need to determine the condition under which a chord's length exceeds the side length of this inscribed equilateral triangle.

```python
import sympy

R = sympy.Symbol('R')

# For an equilateral triangle inscribed in a circle, the side length 's' is given by s = R * sqrt(3).
s = R * sympy.sqrt(3)

print(f"The side length of an equilateral triangle inscribed in a circle of radius R is: s = {s}")
```

### Condition for Chord Length Exceeding Inscribed Triangle Side Length

Let `L` be the length of a chord and `s` be the side length of the inscribed equilateral triangle, which we found to be `s = R * sqrt(3)`. We want to find the condition where `L > s`.

The length of a chord `L` in a circle of radius `R` is related to its distance `d` from the center by the formula: `L = 2 * sqrt(R^2 - d^2)`.

So, the condition `L > s` becomes:
`2 * sqrt(R^2 - d^2) > R * sqrt(3)`

To solve for `d`, we square both sides:
`4 * (R^2 - d^2) > 3 * R^2`
`4R^2 - 4d^2 > 3R^2`
`R^2 > 4d^2`
Taking the square root of both sides (and considering `d` is a non-negative distance):
`R > 2d`
`d < R/2`

Therefore, a randomly selected chord will have a length greater than the side length of the inscribed equilateral triangle if and only if its perpendicular distance from the center of the circle is less than `R/2`.

## Method 1: Random Endpoints on the Circumference

In this method, we select two points, $P_1$ and $P_2$, independently and uniformly at random on the circumference of a circle with radius $R$. The line segment connecting $P_1$ and $P_2$ forms the chord. Our goal is to determine the probability that the length of this chord, $L$, is greater than the side length of an inscribed equilateral triangle, $s = R√3$.

Let's fix one point, $P_1$, at an arbitrary position on the circumference. Without loss of generality, we can place $P_1$ at the top of the circle (e.g., at coordinates $(0, R)$ if the circle is centered at the origin). The position of the second point, $P_2$, can then be described by its angular separation from $P_1$. Let this angular separation be $\theta$, where $\theta$ ranges from $0$ to $2\pi$ radians.

The length of a chord in a circle of radius $R$ is given by the formula $L = 2R \sin(\frac{\theta}{2})$.

We need to find the values of $\theta$ for which $L > s$, i.e., $2R \sin(\frac{\theta}{2}) > R\sqrt{3}$.

Dividing by $2R$ (since $R > 0$):
$\sin(\frac{\theta}{2}) > \frac{\sqrt{3}}{2}$

The general solutions for $\sin(x) = \frac{\sqrt{3}}{2}$ are $x = \frac{\pi}{3}$ and $x = \frac{2\pi}{3}$.

So, we need $\frac{\pi}{3} < \frac{\theta}{2} < \frac{2\pi}{3}$.

Multiplying by 2:
$\frac{2\pi}{3} < \theta < \frac{4\pi}{3}$

This means that for the chord length to be greater than $s$, the angular separation $\theta$ between $P_1$ and $P_2$ must be between $2\pi/3$ (120 degrees) and $4\pi/3$ (240 degrees). The total possible range for $\theta$ is $2\pi$ (from $0$ to $2\pi$).

The favorable range for $\theta$ is $4\pi/3 - 2\pi/3 = 2\pi/3$.

Therefore, the probability is the ratio of the favorable angular range to the total angular range:

$P = \frac{2\pi/3}{2\pi} = \frac{1}{3}$

Thus, for Method 1, the probability that a randomly chosen chord is longer than the side of an inscribed equilateral triangle is $\frac{1}{3}$.

## Method 2: Random Radial Point

In this method, we select a random point along a radius of the circle. A chord is then drawn perpendicular to this radius, passing through the chosen point.

Let `d` be the distance of the chosen point from the center of the circle. The chord's length `L` is related to `R` (radius of the circle) and `d` by the formula `L = 2 * sqrt(R^2 - d^2)`.

From our 'Geometric Setup' phase, we know that a chord's length `L` exceeds the side length of an inscribed equilateral triangle `s = R * sqrt(3)` if its distance from the center `d` is less than `R/2`.

### Probability Calculation for Method 2

For a chord's length `L` to be greater than the side length `s = R * sqrt(3)` of the inscribed equilateral triangle, its distance `d` from the center must satisfy `d < R/2`.

In this method, `d` is chosen randomly along a radius. The radius extends from `d = 0` (the center) to `d = R` (the circumference). Therefore, the possible values for `d` range from `0` to `R`.

The favorable range for `d` is `0 <= d < R/2`.

The total range for `d` is `0 <= d <= R`.

The probability is the ratio of the favorable length of the radius to the total length of the radius:

`Probability = (Favorable length) / (Total length) = (R/2 - 0) / (R - 0) = (R/2) / R = 1/2`

Thus, the probability that a randomly chosen chord (using the random radial point method) has a length greater than the side of the inscribed equilateral triangle is **1/2**.

## Method 3: Random Midpoint

In this method, a random point is chosen uniformly within the entire area of the circle. This chosen point serves as the midpoint of the chord. The chord is then constructed perpendicular to the radius passing through this midpoint. If the chosen midpoint is at the center of the circle, the chord will be a diameter.

### Condition for Chord Length

We previously established that for a chord's length to exceed the side length of an inscribed equilateral triangle, its distance `d` from the center of the circle must be less than `R/2`. The side length of the inscribed equilateral triangle is `s = R * sqrt(3)`. A chord of length `L` at a distance `d` from the center in a circle of radius `R` satisfies the relationship `(L/2)^2 + d^2 = R^2`. For the chord to be longer than `s`, we need `L > R * sqrt(3)`. This implies `(R * sqrt(3) / 2)^2 + d^2 < R^2`, which simplifies to `3R^2/4 + d^2 < R^2`, or `d^2 < R^2/4`, leading to `d < R/2`.

### Determining Favorable Area and Total Sample Space

For a chord's length to be greater than the side length of the inscribed equilateral triangle, its midpoint must lie within a concentric circle of radius `R/2`. This region represents the **favorable area**.

The **total sample space** for the midpoint is the entire circle, which has a radius `R`.

### Calculating the Probability

The probability is the ratio of the favorable area to the total area:

Probability = (Area of inner circle with radius `R/2`) / (Area of outer circle with radius `R`)

Area of a circle = `π * (radius)^2`

```python
import sympy

R = sympy.Symbol('R')

# Area of the outer circle (total sample space)
area_total = sympy.pi * R**2

# Area of the inner circle (favorable area) where the midpoint must lie
# This is a circle with radius R/2
area_favorable = sympy.pi * (R/2)**2

# Calculate the probability
probability = area_favorable / area_total

print(f"The total area for the midpoint (entire circle) is: {area_total}")
print(f"The favorable area for the midpoint (inner circle with radius R/2) is: {area_favorable}")
print(f"The probability that a randomly chosen chord (by midpoint method) is longer than the side of the inscribed equilateral triangle is: {probability}")
```

## Conclusion

We have demonstrated three distinct methods for randomly selecting a chord in a circle, and each method yielded a different probability for the chord's length exceeding the side length of an inscribed equilateral triangle:

*   **Method 1 (Random Endpoints on the Circumference): P = 1/3**
*   **Method 2 (Random Radial Point): P = 1/2**
*   **Method 3 (Random Midpoint): P = 1/4**

This disparity in probabilities, all derived from seemingly reasonable interpretations of "random chord selection," is the essence of Bertrand's Paradox. It illustrates that without a precise definition of the random process, the term "random" can be ambiguous and lead to contradictory results. The paradox highlights the importance of clearly defining the sample space and the probability measure when dealing with continuous random variables.
