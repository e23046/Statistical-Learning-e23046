### Probability Simulation: Two Children, At Least One Girl

This simulation will verify the theoretical probability that both children are girls, given that at least one of them is a girl. We will simulate a large number of families, each with two children, and then count the occurrences of the specified conditions.

```python
import random

def simulate_family():
    """Simulates two children and returns their genders as a tuple."""
    child1 = random.choice(['boy', 'girl'])
    child2 = random.choice(['boy', 'girl'])
    return (child1, child2)

num_simulations = 1000000  # Number of families to simulate

total_at_least_one_girl = 0
total_two_girls_and_at_least_one_girl = 0

for _ in range(num_simulations):
    child1, child2 = simulate_family()

    # Check if at least one child is a girl
    if 'girl' in (child1, child2):
        total_at_least_one_girl += 1
        # If at least one is a girl, check if both are girls
        if child1 == 'girl' and child2 == 'girl':
            total_two_girls_and_at_least_one_girl += 1

# Calculate the conditional probability
if total_at_least_one_girl > 0:
    simulated_probability = total_two_girls_and_at_least_one_girl / total_at_least_one_girl
else:
    simulated_probability = 0 # Should not happen with large num_simulations

print(f"Number of simulations: {num_simulations:,}")
print(f"Families with at least one girl: {total_at_least_one_girl:,}")
print(f"Families with two girls (given at least one girl): {total_two_girls_and_at_least_one_girl:,}")
print(f"Simulated probability: {simulated_probability:.4f}")
print(f"Theoretical probability: {1/3:.4f}")
```
